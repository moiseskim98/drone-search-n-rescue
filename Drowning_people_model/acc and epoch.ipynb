```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files and display their column names to identify the correct ones
train_loss_data = pd.read_csv('train_boxloss.csv')
val_loss_data = pd.read_csv('val_boxloss.csv')
train_acc_data = pd.read_csv('precision.csv')
val_acc_data = pd.read_csv('mAP_0.5.csv')
```

```python
# Extract steps and values using the identified column names
epochs = train_loss_data['Step']  # Assuming 'Step' represents epochs
train_loss = train_loss_data['exp9 - train/box_loss']
val_loss = val_loss_data['exp9 - val/box_loss']
train_accuracy = train_acc_data['exp9 - metrics/precision']
val_accuracy = val_acc_data['exp9 - metrics/mAP_0.5']

# Plot Loss vs Epoch
plt.figure(figsize=(12, 5))

# Training and Validation Loss
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, '-', label='Train Loss')
plt.plot(epochs, val_loss, '-', label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss vs Epoch')

# Training and Validation Accuracy
plt.subplot(1, 2, 2)
plt.plot(epochs, train_accuracy, '-', label='Train Accuracy')
plt.plot(epochs, val_accuracy, '-', label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Accuracy vs Epoch')

plt.tight_layout()
plt.show()
```
